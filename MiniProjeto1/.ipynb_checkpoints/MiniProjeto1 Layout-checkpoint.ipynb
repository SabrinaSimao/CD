### Itens 4 a 6 - Resolução:

```python
SOMA1 = sat.SOMA
soma = SOMA1.dropna()
soma.hist(normed=True, color="m", bins=40)
```

```python
a = s.alpha.fit(soma)
a1,loc,scale = s.alpha.fit(soma)
```

```python
x = np.linspace(800, 2200, 40)
ap = s.alpha.pdf(x, a1,loc,scale)
soma.hist(normed=True, color="m", bins=40)
plt.plot(x, ap, '-', color='green', label="Alpha") 
```

```python
s.lognorm.fit(soma)
s1,loc,scale = s.lognorm.fit(soma)
```

```python
x = np.linspace(800, 2200, 40)
bp = s.lognorm.pdf(x, s1,loc,scale)
soma.hist(normed=True, color="m", bins=40)
plt.plot(x, bp, '-', color='green', label="Alpha") 
```

### Itens 7 a 9 - Resolução:

```python
fra = [] # frequencia relativa acumulada
for i in range(len(soma)):
    acc = (i - 0.5) / len(soma)
    if acc < 0:
        acc = 0
    fra.append(acc)


fra = pd.Series(data=fra)

# Gera os valores dos quantis teóricos para as distribuições analisadas
q_alpha = s.alpha.ppf(fra, 6.1945426418768379, 665.11819171494176, 2857.4473733831719) 
q_lognorm = s.lognorm.ppf(fra, 1.967906528441007, 1100.99999999999977, 150.0895911169255612)


fig = plt.figure(figsize=(12,9))

plt.title("Q-Q plot: Dados x Alpha x Log Normal")

    

exp_plot, = plt.plot(soma, q_alpha, 'o', color="red", label="Alpha")
gamma_plot, = plt.plot(soma, q_lognorm, 'o', color="green", label = "Log Normal")
axes = plt.gca()
axes.set_ylim([0,3000])

plt.legend(handles=[exp_plot, gamma_plot])
plt.plot((min(soma), max(soma)), (min(soma), max(soma)), '-')

for i in range(len(soma)):
    plt.plot((soma[i], soma[i]), (q_alpha[i], soma[i]), '-', color='red', alpha=0.3 )
    plt.plot((soma[i], soma[i]), (q_lognorm[i], soma[i]), '-', color='green', alpha=0.3 )
```
