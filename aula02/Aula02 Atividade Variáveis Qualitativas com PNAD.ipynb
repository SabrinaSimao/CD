```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
notas = pd.read_table("pes2013A.txt", sep=" ")
```

```python
notas["regiao"] = notas.UF



notas.regiao[(notas.UF == 11 )| (notas.UF == 12 )| (notas.UF == 13 )| (notas.UF == 14 )| (notas.UF == 15 )| (notas.UF == 16 )| (notas.UF == 17)] = 'Norte'
notas.regiao[(notas.UF == 21 )| (notas.UF == 22 )| (notas.UF == 23 )| (notas.UF == 24 )| (notas.UF == 25 )| (notas.UF == 26 )| (notas.UF == 27 )| (notas.UF == 28 )| (notas.UF == 29)] = 'Nordeste'
notas.regiao[(notas.UF == 31 )| (notas.UF == 32 )| (notas.UF == 33 )| (notas.UF == 35)] = 'Sudeste'
notas.regiao[(notas.UF == 41 )| (notas.UF == 42 )| (notas.UF == 43)] = 'Sul'
notas.regiao[(notas.UF == 50 )| (notas.UF == 51 )| (notas.UF == 52 )| (notas.UF == 53)] = 'Centro Oeste'
```

```python
"""Aqui copiamos a coluna de UF e a transformamos em uma coluna de regiao"""
notas.regiao
```

```python
notas.shape
notas
```

```python
notas.V4802 = notas.V4802.astype('category')
notas.V4802.cat.categories=('maternal','Alfabetização', 'Alfabetização Adultos', 'Fund 1 JA', 'Fund 2 JA', 'Ensino Fundamental', 'Fund 1', 'Fund 2', 'EJA Modular', 'EM + Cursinho', 'Superior')

notas.V6002 = notas.V6002.astype('category')
notas.V6002.cat.categories= ('Público', 'Privado')

notas.UF = notas.UF.astype('category')
notas.UF.cat.categories=('RR', 'AC', 'AM', 'RO', 'PA', 'AP', 'TO', 'MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA', 'MG', 'ES', 'RJ', 'SP', 'PR', 'SC', 'RS', 'MS', 'MT', 'GO', 'DF')

notas.V0302 = notas.V0302.astype('category')
notas.V0302.cat.categories= ('Masculino', 'Feminino')

notas.V0404 = notas.V0404.astype('category')
notas.V0404.cat.categories= ('Branca', 'Preta', 'Amarela', 'Parda', 'Indígena', 'Sem Declaração')

notas.V4803 = notas.V4803.astype('category')
notas.V4803.cat.categories=('sem instrução', '1 anos','2 anos','3 anos','4 anos','5 anos','6 anos','7 anos','8 anos','9 anos','10 anos','11 anos','12 anos','13 anos','14 anos','15 anos +', 'nao det',)

notas.V06111 = notas.V06111.astype('category')
notas.V06111.cat.categories= ('Tem acesso a internet', 'Não tem acesso')



uf = notas.UF
sex = notas.V0302
cor = notas.V0404
rede = notas.V6002
nivel = notas.V4802
anos_estudo = notas.V4803
internet = notas.V06111
```

```python
tabela = pd.crosstab(uf, [nivel, rede, sex,])
tabela2 = tabela.iloc[:,36:]
tabela2
```

```python
uf_estudo = pd.crosstab(uf, anos_estudo)
uf_estudo
```

```python
"""Esta tabela analisa quantos anos as pessoas estudaram em cada região. O ensino médio completo faz 12 anos de estudo para cada 
aluno. Queremos analisar as pessoas que tem 12 anos ou mais de estudo, pois estas indicam que seu local é positivo para 
instalação de uma nova faculdade"""

uf_estudoCut = uf_estudo.iloc[:,12:16]
uf_estudoCut
```

```python

uf_estudoCut.sum(axis=1)
```

```python
tabela2.plot.bar(color=('yellow', 'salmon', 'blue', 'red', 'green','black','pink','orange'), stacked=True, figsize=(10,10))

tabelaSP = tabela.iloc[19:20,36:]
tabelaSP.plot.bar(color=('yellow', 'salmon', 'blue', 'red', 'green','black','pink','orange'), stacked=True, legend=False)
```

```python
uf_nivel = pd.crosstab(uf, nivel)
uf_nivel
```

```python
uf_nivelCut = uf_nivel.iloc[:,9:]
uf_nivelCut
```

```python
uf_nivelCut.plot.bar(color=('yellow', 'black'), stacked=False, figsize=(10,10))
uf_nivelCut.plot.bar(color=('yellow', 'black'), stacked=True, figsize=(10,10))
```

```python
uf_rede = pd.crosstab(uf, rede)
uf_redecut = uf_rede.iloc[:,:]
uf_redecut
```

```python
uf_redecut.plot.bar(color=('pink', 'black'), stacked=False, figsize=(10,10))
```

```python
uf_internet = pd.crosstab(uf, internet)
uf_internet
```

```python
uf_internetCut = uf_internet.iloc[:,:1]
```

```python
uf_internetCut.plot(kind='pie', colors=('pink', 'purple'), autopct="%0.2f",legend=False, subplots=True, figsize=(11,11))
```

```python
"""Concluindo, os estados com maior preferencia para construção e desenvolvimeto de uma nova faculdade particular são os
estados do sudeste, pois aparentam em todas as medidas valores superiores de: formandos do EM, cursadores de curso superior,
cursadores de escolas particulares, acesso a internet, e taxa de ensino para adultos. Apesar disso, se o tal aclamado Sr. Gold 
quiser se aventurar e instalar uma faculdade em um local com menor concorrência, sua melhor opção seria em um local com grande
taxa de conclusão de EM e pouca adesão ao curso superior (por falta de opções locais). 
Nesse caso, seriam locais bons: Amapá, Para, Bahia"""
```
